```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path='/kaggle/input/ushape/ushape.csv'
df=pd.read_csv(path)
```

```python
headers = ['X', 'Y', 'class']

# Read the CSV file into a DataFrame and add headers
df = pd.read_csv(path, header=None, names=headers)

# Displaying the DataFrame
print(df.head())
```

```python
plt.scatter(df['X'],df['Y'],c=df['class'])
```

```python
X = df[['X', 'Y']].values
y = df['class'].values

# Displaying the extracted data
print("X:")
print(X)
print("\ny:")
print(y)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

# Using HE Initializtion

```python
model=Sequential()
model.add(Dense(10,activation='relu',input_dim=2,kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
```

# Using Glorat/Xavier Initialization

```python
model1 = Sequential()
model1.add(Dense(10, activation='tanh', input_dim=2, kernel_initializer='glorot_uniform'))
model1.add(Dense(10, activation='tanh', kernel_initializer='glorot_uniform'))
model1.add(Dense(10, activation='tanh', kernel_initializer='glorot_uniform'))
model1.add(Dense(10, activation='tanh', kernel_initializer='glorot_uniform'))
model1.add(Dense(1, activation='sigmoid'))
```

```python
model1.compile(loss='binary_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=100,validation_split=0.2)
```
