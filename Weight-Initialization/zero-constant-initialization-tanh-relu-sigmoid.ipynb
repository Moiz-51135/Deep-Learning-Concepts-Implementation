```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path='/kaggle/input/ushape/ushape.csv'
df=pd.read_csv(path)
```

```python
df.head()
```

```python
headers = ['X', 'Y', 'class']

# Read the CSV file into a DataFrame and add headers
df = pd.read_csv(path, header=None, names=headers)

# Displaying the DataFrame
print(df.head())
```

```python
plt.scatter(df['X'],df['Y'],c=df['class'])
```

```python
X = df[['X', 'Y']].values
y = df['class'].values

# Displaying the extracted data
print("X:")
print(X)
print("\ny:")
print(y)
```

```python
import tensorflow
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

# Using relu and initializing weights to Zero

```python
model=Sequential()
model.add(Dense(12,activation='relu',input_dim=2))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
# Set parameters to 0
model.get_weights()
```

```python
initial_weights = model.get_weights()
```

```python
initial_weights[0] = np.zeros(model.get_weights()[0].shape)
initial_weights[1] = np.zeros(model.get_weights()[1].shape)
initial_weights[2] = np.zeros(model.get_weights()[2].shape)
initial_weights[3] = np.zeros(model.get_weights()[3].shape)
```

```python
model.set_weights(initial_weights)
```

```python
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(X,y,epochs=100,validation_split=0.2)
```

```python
model.get_weights()
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X,y.astype('int'), clf=model, legend=2)
```
