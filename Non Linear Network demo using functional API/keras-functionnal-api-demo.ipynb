```python
from keras.models import Model
```

```python
model = Model(inputs = x ,outputs = [output_1,output_2])
```

```python
from keras.layers import *
```

```python
x = Input(shape=(3,))
```

```python
hidden_layer1=Dense(128,activation='relu')(x)
hidden_layer2=Dense(64,activation='relu')(hidden_layer1)

output_1=Dense(1,activation='linear')(hidden_layer2)
output_2=Dense(1,activation='sigmoid')(hidden_layer2)
```
