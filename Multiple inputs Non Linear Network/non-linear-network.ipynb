```python
from keras.layers import *
from keras.models import Model
```

```python
Input_1=Input(shape=(32,))
Input_2=Input(shape=(64,))
```

# Creating Branches for Input_1

```python
x1=Dense(128,activation='relu')(Input_1)
x2=Dense(64,activation='relu')(x1)
```

# Creating Branches for Input_2

```python
y1=Dense(32,activation='relu')(Input_2)
y2=Dense(16,activation='relu')(y1)
y3=Dense(8,activation='relu')(y2)
```

# Concatenating/Combining outputs of two declared inputs

```python
concatenation=concatenate([x2,y3])
```

# Connecting with Fully connected layers and Ouput node

```python
z=Dense(2,activation='relu')(concatenation)
out=Dense(1,activation='linear')(z)
```

```python
model=Model(inputs=[Input_1,Input_2],outputs=[z,out])
```

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model)
```
